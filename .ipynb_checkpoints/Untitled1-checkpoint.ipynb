Моё понимание. Мы имеем сгенерированные p и g числа, которые знают все. У каждого пользователя имеется свой секретный ключ. В данном случае а - Алиса и b - Боб. Мы подготовливаем специальное зашифрованное сообщение ввиде числа А у Алисы. Боб, получив число А возводит его в свой секретный ключ и получает нужный остаток. 

```python
def gcd(a,b):
    while a != b:
        if a > b:
            a = a - b
        else:
            b = b - a
    return a

def prm(m):
    required_set = set(num for num in range (1, m) if gcd(num, m) == 1)
    for g in range(1, m):
        actual_set = set(pow(g, powers) % m for powers in range (1, m))
        if required_set == actual_set:
            print(g)          
```

```python
p = 23
prm(p)

g = input("Выберите g из списка выше: ")

a = 6
# Открытый ключ A.
A = g ** a % p
print(A)

b = 4
# Открытый ключ B.
B = g ** b % p
K = A ** b % p
print(K)

K = B ** a % p
print(K)
```
